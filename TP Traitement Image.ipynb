## 1. Lire et écrire un pixel

Voici notre image de départ.

On précise que le repérage sur l’image se fait selon un repère d’origine le coin supérieur gauche de l’image, d’axe des abscisses horizontal orienté vers la droite et d’axe des ordonnées vertical orienté vers le bas.
Nous allons voir comment lire le pixel de coordonnées (100,250). 

1. Exécuter le script ci-dessous et noter le résultat.

```python
from PIL import Image
path="C:/Users/Laurine/Pictures/jupyter/Pomme.jpg"
img=Image.open(path)
r,v,b=img.getpixel((100,250))
print("canal rouge : ",r,", canal vert : ",v,", canal bleu : ",b)
```
